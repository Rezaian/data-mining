```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Dataset Parameters
np.random.seed(42)
m = 100  # Number of examples
n = 2    # Number of features

# Generate random features
X = np.random.rand(m, n)

# Define true parameters for generating target
true_theta = np.array([3, 2, 1.5]).reshape(-1, 1)  # True [theta0, theta1, theta2]
noise = np.random.randn(m, 1)  # Gaussian noise

# Generate target variable
y = true_theta[0] + true_theta[1] * X[:, 0].reshape(-1, 1) + true_theta[2] * X[:, 1].reshape(-1, 1) + noise

# Add bias term (x0 = 1) to the feature matrix
X_b = np.c_[np.ones((m, 1)), X]

# print("Dataset generated:")
# print("First 5 rows of X:\n", X[:5])
# print("First 5 rows of y:\n", y[:5])
```

```python
# Plotting x1 vs y
plt.figure(figsize=(12, 5))

# Plotting x0 vs y

plt.subplot(1, 3, 1)
plt.scatter(X_b[:, 0], y, color='blue', alpha=0.6)
plt.xlabel('$x_1$')
plt.ylabel('$y$')
plt.title('Feature $x_0$ vs Target $y$')

# Plotting x1 vs y
plt.subplot(1, 3, 2)
plt.scatter(X_b[:, 1], y, color='green', alpha=0.6)
plt.xlabel('$x_2$')
plt.ylabel('$y$')
plt.title('Feature $x_1$ vs Target $y$')


# Plotting x2 vs y
plt.subplot(1, 3, 3)
plt.scatter(X_b[:, 2], y, color='green', alpha=0.6)
plt.xlabel('$x_2$')
plt.ylabel('$y$')
plt.title('Feature $x_2$ vs Target $y$')

plt.tight_layout()
plt.show()
```

```python
# Initialize parameters randomly
theta = np.random.randn(n + 1, 1)  # Random initial theta [theta0, theta1, theta2]

# Hyperparameters
learning_rate = 0.1
iterations = 1000

print("Initial Parameters (theta):\n", theta)
print(f"Learning rate: {learning_rate}, Iterations: {iterations}")
```

```python
# Gradient Descent
m = len(y)  # Number of training examples
cost_history = []

for iteration in range(iterations):
    predictions = X_b.dot(theta)  # Hypothesis
    errors = predictions - y      # Errors
    gradients = (1 / m) * X_b.T.dot(errors)  # Gradient computation
    theta -= learning_rate * gradients       # Update theta
    cost = (1 / (2 * m)) * np.sum(errors**2)  # Compute cost (MSE)
    cost_history.append(cost)

    # Print cost every 200 iterations for progress tracking
    if iteration % 200 == 0:
        print(f"Iteration {iteration}: Cost = {cost:.4f}")

print("\nGradient Descent Complete.")
print("Learned Parameters (theta):\n", theta)
```

```python
# Plot cost function over iterations
plt.figure(figsize=(8, 5))
plt.plot(range(len(cost_history)), cost_history, color='blue')
plt.xlabel('Iterations')
plt.ylabel('Cost J(theta)')
plt.title('Cost Function Over Iterations')
plt.grid(True)
plt.show()
```

```python
# Compare learned and true parameters
print("True Parameters:")
print(true_theta)

print("\nLearned Parameters:")
print(theta)
```
